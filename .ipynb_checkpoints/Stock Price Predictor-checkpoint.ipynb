```python
import numpy as np
import pandas as pd
import yfinance as yf
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
stocks = input('Enter the code of stock :')
data = yf.download(stocks,'2008-01-01','2022-01-07', auto_adjust=True)
data.head()
```

```python
# Things we learned

#1. the shape of our data
#2. how our data is distributed
#3. our data is very non-linear

#4. linear, logistic, regualrized, SVM
```

```python
X = data.drop('Close',axis=1)
y = data["Close"]
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train,y_train)
pred1 = lr.predict(X_test)
```

```python
pred1
```

```python
y_test
```

```python
# MSE - mean squared error
# RMSE - square root of mean squared error
# r2_square -- 1.0 #good model
```

```python
from sklearn.metrics import mean_squared_error, r2_score

def calculate_metrics(y_test,y_pred):
    '''y test is ground truth
    y_pred is model pred value'''
    
    mse = mean_squared_error(y_test,y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test,y_pred)
    
    print(mse,rmse,r2)
    
calculate_metrics(y_test,pred1)
    
    
```

```python
#using regularized models

from sklearn.linear_model import Lasso,Ridge

la = Lasso()
la.fit(X_train,y_train)
ri = Ridge()
ri.fit(X_train,y_train)

lap = la.predict(X_test)
rip = ri.predict(X_test)
```

```python
calculate_metrics(y_test,lap)
```

```python
calculate_metrics(y_test,rip)
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV

svr = SVR()

param_grid = {'C':[0.1,1,10,100,1000],
             'gamma':[1,0.1,0.01,0.001,0.0001],
             'kernel':['rbf']}

grid = GridSearchCV(SVR(),param_grid,refit=True,verbose=3)
grid.fit(X_train,y_train)
```

```python
svr = SVR(C=10,gamma=0.01,kernel='rbf')
svr.fit(X_train,y_train)
svr_pred = svr.predict(X_test)
```
